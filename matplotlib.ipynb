```python
import matplotlib.pyplot as plt
import numpy as np
import random
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

random.seed(42)

x = [random.uniform(0, 10) for _ in range(1000)]
y = [xi * 2 + random.gauss(0, 2) for xi in x]

# 计算线性回归
slope, intercept = np.polyfit(x, y, 1)

plt.figure(figsize=(10, 6))

plt.scatter(
    x, y,
    color='blue',
    alpha=0.5,
    s=20,
    edgecolors='none',
    label='数据点'
)

plt.plot(
    [0, 10],
    [intercept, slope * 10 + intercept],
    color='red',
    linestyle='--',
    linewidth=2,
    label=f'回归线: y = {slope:.2f}x + {intercept:.2f}'
)

plt.title('1000个随机点的散点图', fontsize=14)
plt.xlabel('X轴：自变量', fontsize=12)
plt.ylabel('Y轴：因变量', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.tight_layout()

plt.show()
```
